```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
myretaildata = pd.read_excel("Online_Retail.xlsx")
myretaildata.head()
```

```python
#data cleaning
myretaildata['Description'] = myretaildata['Description'].str.strip() #removes spaces from beginning and end
myretaildata.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
myretaildata['InvoiceNo'] = myretaildata['InvoiceNo'].astype('str') #to convert to string
myretaildata = myretaildata[~myretaildata['InvoiceNo'].str.contains('C')]#remove the credit transactions  (example at index 141)
myretaildata.head()
```

```python
myretaildata['Country'].value_counts()
```

```python
# seperating them according to country
my_basket = myretaildata[myretaildata['Country']=='Germany'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
#yaha par sum ka kaam acche se smjh nhi aaya lekin agr step by step cheezen karein to phr agr hum sirf ['Quantity'] likhenge to phr vo hume bss Quantity ka data de dega lekin phr jab hum usse sum karenge to unn groups ke data ko ek sath dikayega
```

```python
my_basket.first() #this is used to see how they are grouped though they do not appear in .head() call
```

```python
my_basket.head()
```

```python
#converting all positive vaues to 1 and everything else to 0
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

my_basket_sets = my_basket.applymap(encode_units)
my_basket_sets.head()
```

```python
some_column = pd.DataFrame(my_basket_sets.columns)
```

```python
some_column[some_column["Description"] == "POSTAGE"] #vese mujhe to pata tha column ka naam isliye search kr liya varna code krte krte pata chalta hai
```

```python
my_basket_sets.drop('POSTAGE', inplace=True, axis=1) #Remove "postage" as an item
```

```python
#generating frequent itemsets
my_frequent_itemsets = apriori(my_basket_sets, min_support=0.05, use_colnames=True)
```

```python
#generating rules
my_rules = association_rules(my_frequent_itemsets, metric='lift', min_threshold=1)
```
